```python
import os
import matplotlib.pyplot as plt
import numpy as np

from src.backstage.evalutils import Evaluator
```

```python
def showScatterPlot(k1, error_k1, name):
    params = {
            'font.size': 20,
            'legend.loc': 'upper right'
        }
    plt.rcParams.update(params)
    fig = plt.figure(figsize=(14,8))
    axes = fig.subplots(1, 1)
    axes.set_ylim(top=0.15, bottom=-0.15)
    plt.scatter(k1, error_k1, s=1.0, label=name)
    plt.axhline(y=0.0, color="r", linestyle="--", linewidth=2.0)
    axes.set_xlabel("$k_1$")
    axes.set_ylabel("Error of $\widehat{k_1}$")
    plt.legend()
    plt.show()
    plt.close()
```

```python
BASE_CONFIG_PATH="config"

configs = [
    ("seta-densenet161.yaml", "DenseNet161 - Set A"),
    #("setb-densenet161.yaml", "DenseNet161 - Set B"),
    ("setc-densenet161.yaml", "DenseNet161 - Set C"),
    #("seta-resnet152.yaml", "ResNet152 - Set A"),
    #("setb-resnet152.yaml", "ResNet152 - Set B"),
    ("setc-resnet152.yaml", "ResNet152 - Set C"),
    #("seta-efficientnetb5.yaml", "EfficientNet-B5 - Set A"),
    #("setb-efficientnetb5.yaml", "EfficientNet-B5 - Set B"),
    ("setc-efficientnetb5.yaml", "EfficientNet-B5 - Set C")
]

for (config,label) in configs:
    fn = os.path.join(BASE_CONFIG_PATH, config)
    ev = Evaluator(configFile=fn)   
    k1, error_k1 = ev.evaluate()

    showScatterPlot(k1, error_k1, label)
```
